```python
# Kacper Skelnik, 291566

import matplotlib.pyplot as plt
import numpy as np
```

```python
def funkcja(x):
    return 3*x**3 - 4*x**2 - 1
```

```python
def bisekcja(a, b, epsilon, tab):
    
    if a>b:
        a,b = b,a

    if funkcja(a)*funkcja(b)<0:
        pierwiastek = a
        while abs(b-a) >= abs(epsilon):
            pierwiastek = (b + a) / 2.0 
            tab.append(pierwiastek)
            print(pierwiastek)
            if funkcja(a)*funkcja(pierwiastek)<0:
                b = pierwiastek
            elif funkcja(pierwiastek)*funkcja(b)<0:
                a = pierwiastek
            else:
                break

    elif abs(funkcja(a)) <= epsilon:
        pierwiastek = a
    elif abs(funkcja(b)) <= epsilon:
        pierwiastek = b
    else:
        pierwiastek = "Blad - podany przedzial nie spelnia zalozen" 
       
    return pierwiastek

def wykres(a,b,tabx):
    taby = []
    for i in range(len(tabx)):
        taby.append(funkcja(tabx[i]))

    x = np.linspace(a, b, 10000)
    plt.plot(x, funkcja(x), label='funkcja')
    plt.plot(tabx, taby, 'o')
    plt.axhline(y=0, color='black')
    plt.show()
    
def program(a, b, epsilon):
    tabx = []
    print("rozwiazanie = ", bisekcja(a, b, epsilon, tabx))
    wykres(a,b,tabx)
```

```python
a = -4
b = 4

#program(punkt a, punkt b, epsilon)

program(a, b, 0.0001)
```
