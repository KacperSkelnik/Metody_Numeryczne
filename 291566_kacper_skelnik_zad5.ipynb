```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([52,5,-5,-40,10])
```

```python
def dopasuj(x,y,n):
    if(len(x) == len(y) and n < len(x)):
    
        xsum = np.zeros(2*n)
        A = np.zeros((n, n))
        b = np.zeros(n)

        for i in range(0, 2*n):
            xsum[i] = np.sum(x**i)
    
        for i in range(0, n):
            for j in range(0, n):
                A[i][j] = xsum[i+j]
            b[i] = np.sum((x**i)*y)
        
        a = np.linalg.solve(A,b)
        a = np.flip(a)
        return a
    else:
        print("wybrany stopien wielomianu > n - 1")
        return None
```

```python
n = 3 #stopie≈Ñ wielomianu
a = dopasuj(x,y,n)

x_poly = np.linspace(1,5,100)
array = np.polyval(a,x_poly)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(x,y, color = 'r')
plt.plot(x_poly,array)

plt.title("Dopasowana prosta n = 3", size = 15)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)

plt.show()
```
