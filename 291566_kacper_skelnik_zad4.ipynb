```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
macierz = np.array([[6,5,-5],[2,6,-2],[2,5,-1]])
wektor = np.array([random.random(),random.random(),random.random()])
```

```python
y = np.array([0,0,0])
wartosci_wlasne = []
x = list(range(0, 30))

for i in range(30):
    y = wektor/np.linalg.norm(wektor)
    wartosci_wlasne.append(np.linalg.norm(wektor))
    wektor = np.dot(macierz, y)
```

```python
w, v = np.linalg.eig(macierz)

np.set_printoptions(precision=3)

print("Korzystając z funkcji wbudowanej numpy.linalg.eig zostały obliczone wartości i wektory własne: \n")
print("pierwsza wartość własna: {:.2f}, odpowiadający jej wektor własny:".format(w[0]), v[:,0])
print("druga wartość własna: {:.2f}, odpowiadający jej wektor własny:".format(w[1]), v[:,1])
print("trzecia wartość własna: {:.2f}, odpowiadający jej wektor własny:".format(w[2]), v[:,2], "\n")
print("Wartość i wektor własny obliczony przeze mnie: \n")
print("Wartość własna: {:.2f}, wektor własny: ".format(wartosci_wlasne[-1]), y, "\n")
```

```python
plt.figure(figsize=(8,6))

plt.plot(x,wartosci_wlasne)
plt.title("Zależność między wartością własną a ilością iteracji", size = 15)
plt.xlabel("iteracje", size = 14)
plt.ylabel("Wartość własna", size = 14)

plt.show()
```
